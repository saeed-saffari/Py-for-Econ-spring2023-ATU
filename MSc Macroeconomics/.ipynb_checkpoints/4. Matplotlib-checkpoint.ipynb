## Import

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Specification

```python
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 50)
y = x ** 2

plt.figure(figsize=(10,5))
plt.plot(x, y,
         c = 'r', ls = '-', lw = 3,
         marker = 'o', markersize = 10, markerfacecolor = 'm', markeredgecolor = 'b',
         label = '$Y = X^2$'
        )

plt.legend(loc = 'upper center', fontsize = 9)
plt.xlabel('X range', fontsize = 10, fontweight = 2, fontname = 'Times New Roman')
plt.ylabel('Y range', fontsize = 10, fontweight = 2, fontname = 'Times New Roman')
plt.title('Function $Y = x^2$ \nin range (-10,10)', fontsize = 15, fontweight = 2, fontname = 'Times New Roman', color = 'darkblue')

plt.grid(ls = '--', c = 'k', alpha = 0.6)

plt.axhline(40, lw = 2, c = 'darkorange', ls = '-')
plt.axvline(0, lw = 2, c = 'darkgreen', ls = '-' )

#plt.xlim(left = -2.5, right = 8)
#plt.ylim((20, 80))

plt.text(-5, 60, 'First text')

#plt.savefig('test_1.jpg', dpi = 700)

plt.show()
```

```python
x = np.linspace(-10, 10, 50)

plt.figure(figsize=(10,5))
#z = 150*np.sin(x)

plt.plot(x, 10 * x**2, c = 'r', ls = '-', lw = 3, label = '$y = x^2$')
plt.plot(x, x**3, c = 'b', ls = '--', lw = 3, label = '$y = x^3$')
plt.plot(x, 200*np.sin(x), c = 'g', ls = ':', lw = 3, label = '$y = sin(x)')

plt.legend()
plt.grid(ls = '--', c = 'k', alpha = 0.6)


plt.show()
```

## Bar

```python
plt.figure(figsize=(10,5))

plt.bar([1,3,5,7], [5,8,2,6], label = 'ONE', color = 'g')
plt.bar([2,4,6,8], [1,4,8,3], label = 'TWO', color = 'r')

plt.grid(ls = '--', c = 'k', axis = 'y', alpha = 0.5)
plt.legend()
plt.show()
```

## Scatter

```python
x = np.linspace(0,10, 100)
e = np.random.rand(100) * 100
y = x + e

scale = np.random.rand(100) * 500
color = np.random.rand(100) * 10

plt.figure(figsize=(10,5))
plt.scatter(x, y, s = scale, c = color)

plt.colorbar().set_label('Color bar', fontsize = 12)

plt.grid()
plt.show()
```

```python
z = np.arange(100)
scale = np.arange(100) * 2
color = np.arange(100)
e = np.random.randn(100)
y = z + e * 20

plt.figure(figsize=(10,5))
plt.scatter(z, y, c = color, s = scale)
plt.colorbar()
plt.grid()
plt.show()
```

## Histogram

<img stc = https://wikimedia.org/api/rest_v1/media/math/render/svg/00cb9b2c9b866378626bcfa45c86a6de2f2b2e40 >

```python
data = np.random.normal(0, 1, 1000000)

x = np.linspace(-4,4, 100)
y = 1 / np.sqrt(2* np.pi) * np.exp(-x**2/2)

plt.figure(figsize=(10,5))
plt.plot(x, y, lw = 3, c = 'k')
plt.hist(data, 700, density=1)
plt.show()
```
